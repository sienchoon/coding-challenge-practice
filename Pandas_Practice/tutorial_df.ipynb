## Filter with conditional logic using `loc[]`

```python
for i in df[["Name", "Platform", "Year"]]:
    print(df[i].unique())
```

```python
PlayStation4 = df["Platform"] == "PS4"
PlayStation4.value_counts()
```

```python
df.loc[PlayStation4]
```

```python
# df.locate [ condition ] where condition are the column of dataframes, and shows the "Name" as results
df.loc[df["JP_Sales"] > df["NA_Sales"], "Name"]
```

```python
df.loc[(df["Platform"] == "PS4") & (df["Genre"] == "Action")].sort_values(by = "Global Sales", ascending = False)
```

```python
rpg = df["Genre"]  == "Role-Playing"
df.loc[rpg & PlayStation4].head()
```

```python
#Filter with OR

action = df["Genre"] == "Action"
shooter = df["Genre"] == "Shooter"

df.loc[action | shooter]
```

```python
since_2010 = df["Year"] >= 2010

df.loc[(since_2010) & (action | shooter)]
```

```python
top_5_NA_Sales = df.sort_values(by="NA_Sales", ascending = False)
top_5_NA_Sales.head()
```

```python
df["Genre"].unique()
df["Genre"].value_counts()
```

```python
Shooter = df["Genre"] == "Shooter"
year_2010 = df["Year"] == 2010
df.loc[Shooter & (year_2010)]

```

```python
# isin() method
# .isin([ put in the list of elements])
# df["Publisher"] == df.Publisher
top_3_publisher =  df.Publisher.isin(["Electronic Arts", "Activision", "Namco Bandai Games"])
df.loc[top_3_publisher]
```

```python
genres = df["Genre"].isin(["Action", "Shooter"])
sales_2M = df["Global Sales"] >= 2
df.loc[genres & sales_2M & top_3_publisher]
```

```python
#between
# allows to specify a range
between_2M_10M = df["Global Sales"].between(2,10)

year_90s = df["Year"].between(1990, 1999)
year_00s = df["Year"].between(2000, 2009)
year_10s = df["Year"].between(2010, 2019)
```

```python
df.loc[between_2M_10M & year_90s, "Genre"].value_counts(normalize=True)
```

## `GroupBy` function

```python
# read from right to left
# average JP sales by the genre
df.groupby("Genre")["JP_Sales"].mean()
```

```python
# locate the average sales in Japan in the Shooter genre

df.loc[df["Genre"] == "Shooter", "JP_Sales"].mean()
```

```python
sales_by_genre = df.groupby("Genre")[["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales", "Global Sales"]].mean()
sales_by_genre
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(sales_by_genre, annot=True, cmap="Greens")
```

```python
df.groupby("Publisher")["NA_Sales"].agg(["mean", "median", "count"])
```
