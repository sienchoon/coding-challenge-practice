```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
csv_path = 'November 2023 Sales Breakdown.csv'


sales_df = pd.read_csv(csv_path)
sales_df.head(10)
```

```python
drop_columns = ['LocationNumber', 'LocationName']
sales_df = sales_df.drop(columns=drop_columns, errors='ignore')
sales_df.head(10)
```

```python
coffee_df = sales_df[sales_df['Group1'] == 'Coffee']
coffee_df
```

```python
coffee_df.groupby('Description')['Quantity'].max()
```

```python
extracted_text = coffee_df['Description'].str.extract(r'(\b\w+(?:(?<!\s)(?!T/A|D/I)\s?\w*)*\b) (\w+)', expand=True)
extracted_text.columns = ['Drink Type', 'Drink Size']

coffee_type = pd.concat([coffee_df, extracted_text], axis=1)
coffee_type
```

```python
custom_order = ['Sml', 'Med', 'Lrg', 'Cup', 'Mug', 'Glass']
coffee_type['Drink Size'] = pd.Categorical(coffee_type['Drink Size'], categories=custom_order, ordered=True)
coffee_type.sort_values(by=['Drink Type', 'Drink Size'], ascending=True)
```

```python
Cappuccino = coffee_type.groupby(['Drink Type', 'Drink Size']).size().reset_index(name='Counts')
print(Cappuccino) #Cappuccino
```

```python
fig, ax = plt.subplots(figsize=(8, 6))  
x = Cappuccino
colors = plt.get_cmap('Blues')(np.linspace(0.2, 0.7, len(x)))
ax.pie(x, radius=3, colors=colors, startangle=90, wedgeprops={'linewidth': 1.5, 'edgecolor': 'white'},frame=True)
```
