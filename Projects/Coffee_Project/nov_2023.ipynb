```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
csv_path = 'nov_2023_sales.csv'


coffee_df = pd.read_csv(csv_path)
coffee_df.head(25)
```

```python
drop_columns = ['LocationNumber', 'LocationName', 'Chart1_SeriesGroup_label']
coffee_df = coffee_df.drop(columns=drop_columns, errors='ignore')
coffee_df.head()
```

```python
coffee_rename = coffee_df.rename(columns={
    'Chart1_SeriesGroup_Chart1_CategoryGroup_label': 'Group',
    'Chart1_SeriesGroup_Chart1_CategoryGroup_Value_Y': 'Total Amount',
    'Textbox13': 'Quantity',
    'txtTotalSales': 'Total Sales'
})
coffee_rename.head()
    
```

```python
coffee_rename.info()
```

```python
coffee_rename['Quantity'] = coffee_rename['Quantity'].str.replace(',', '').astype(float)
coffee_rename['Total Sales'] = coffee_rename['Total Sales'].str.replace(',', '').astype(float)
```
